```python
# Bibliotecas base
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Configurações gerais do Pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Leitura do arquivo 'sales_data_sample.csv'
df = pd.read_csv('sales_data_sample.csv', encoding='utf-8', index_col=0)
df.head()
```

```python
df.info()
```

```python
# Converte a coluna 'ORDERDATE' para o formato 'DATATIME'
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
# Apaga as colunas 'MONTH_ID' e 'YEAR_ID'
df.drop(['MONTH_ID', 'YEAR_ID'], axis=1, inplace=True)
```

```python
# Avaliar Endereço 1
df['ADDRESSLINE1'].isnull().value_counts()
# Não há valores nulos em Endereço 1
```

```python
# Avaliar Endereço 2
df['ADDRESSLINE2'].isnull().value_counts()
# Existem muitos valores nulos em Endereço 2
```

```python
# Apaga a coluna 'ADDRESSLINE2'
df.drop(['ADDRESSLINE2'], axis=1, inplace=True)
```

```python
# Avalia Território
df['TERRITORY'].value_counts()
```

```python
# Avaliar 'PHONE'
df['PHONE']
```

```python
# Apaga a coluna 'PHONE'
df.drop(['PHONE'], axis=1, inplace=True)
```

```python
# Apaga a coluna 'ADDRESSLINE1', 'POSTALCODE', 'TERRITORY', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME'
df.drop(['ADDRESSLINE1', 'POSTALCODE', 'TERRITORY', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME'], axis=1, inplace=True)
```

```python
# Apaga 'ORDERLINENUMBER'
df.drop(['ORDERLINENUMBER'], axis=1, inplace=True)
```

```python
df['PRODUCTCODE'].nunique()
```

```python
df.rename(columns={
    'QUANTITYORDERED': 'Encomenda',
    'PRICEEACH': 'Valor (US$)',
    'SALES': 'Venda',
    'ORDERDATE': 'Data Pedido',
    'STATUS': 'Status',
    'QTR_ID': 'Trimestre',
    'PRODUCTLINE': 'Linha de Produto',
    'MSRP': 'Varejo',
    'PRODUCTCODE': 'Código Produto',
    'CUSTOMERNAME': 'Cliente',
    'CITY': 'Cidade',
    'STATE': 'Estado',
    'COUNTRY': 'País',
    'DEALSIZE': 'Tamanho do Negócio'
    }, inplace=True)
```

```python
df['customer_id'].nunique()
```

```python
df['Status'].unique()
```

```python
df.loc[df['Status'] == 'Shipped', 'Status'] = 'Enviado'
df.loc[df['Status'] == 'Disputed', 'Status'] = 'Em negociação'
df.loc[df['Status'] == 'In Process', 'Status'] = 'Em processamento'
df.loc[df['Status'] == 'Cancelled', 'Status'] = 'Cancelado'
df.loc[df['Status'] == 'On Hold', 'Status'] = 'Em espera'
df.loc[df['Status'] == 'Resolved', 'Status'] = 'Resolvido'

df['Status'].unique()
```

```python
fig, ax = plt.subplots(figsize=(9,4))
ax.pie(df.groupby('Status')['Encomenda'].nunique(), labels=df['Status'].unique(), autopct='%1.2f%%')
fig.suptitle("Distribuição por Status")
fig.legend(df['Status'].unique(), loc='center right', ncols=1);
```
